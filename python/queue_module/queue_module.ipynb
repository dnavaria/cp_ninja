```python
class custom_queue:
    def __init__(self):
        self.__c_queue = []
        self.__front_ptr = 0
        self.__rear_ptr = 0

    def enqueue(self,item):
        self.__c_queue.append(item)
        self.__rear_ptr+=1
    
    def dequeue(self):
        if self.__rear_ptr == 0:
            return None
        data = self.__c_queue[self.__front_ptr]
        self.__front_ptr+=1
        self.__rear_ptr-=1
        return data

    def front(self):
        if self.isEmpty():
            return None
        return self.__c_queue[self.__front_ptr]

    def size(self):
        return self.__rear_ptr

    def isEmpty(self):
        return self.size() == 0

    def __repr__(self):
        return f"<custom queue :: {self.__c_queue}>"

cq = custom_queue()
print(cq.front())
cq.enqueue(1)
cq.enqueue(2)
cq.enqueue(3)
print(cq.front())
while cq.isEmpty() == False:
    print(cq)
    print(cq.dequeue())
print(cq.front())
```
